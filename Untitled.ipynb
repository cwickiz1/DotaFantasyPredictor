```python
recent_1000 = match_df.match_id.unique()[:1000]
```

```python
KILLS = 0.3
ASSISTS = 0.15
DEATHS_FLAT = 3
DEATHS_MULT = -0.3
LAST_HITS = 0.003
GPM = 0.002
ROSHAN = 1
TOWERS = 1
TEAMFIGHTS = 3
FB = 4
STUNS = 0.05
RUNES = 0.25
WARDS = 0.5
STACKS = 0.5
```

```python
def make_fantasy(series):
    ret = pd.Series(dtype='float64')
    ret['kill_score'] = series['kills'] * KILLS
    ret['assist_score'] = series['assists'] * ASSISTS
    ret['death_score'] = (DEATHS_FLAT + series['deaths'] * DEATHS_MULT)
    ret['lasthit_score'] = (series['last_hits'] + series['denies']) * LAST_HITS
    ret['gpm_score'] = series['gold_per_min'] * GPM
    ret['fb_score'] = series['firstblood_claimed'] * FB
    ret['tower_score'] = series['towers_killed'] * TOWERS
    ret['stack_score'] = series['camps_stacked'] * STACKS
    ret['roshan_score'] = series['roshans_killed'] * ROSHAN
    ret['stun_score'] = series['stuns'] * STUNS
    ret['teamfight_score'] = series['teamfight_participation'] * TEAMFIGHTS
    ret['obs_score'] = series['obs_placed'] * WARDS
    ret['rune_score'] = series['rune_pickups'] * RUNES
    ret['fantasy_score'] = ret['kill_score'] + \
                            ret['assist_score'] + \
                            ret['death_score'] + \
                            ret['lasthit_score'] + \
                            ret['gpm_score'] + \
                            ret['fb_score'] + \
                            ret['tower_score'] + \
                            ret['stack_score'] + \
                            ret['roshan_score'] + \
                            ret['stun_score'] + \
                            ret['teamfight_score'] + \
                            ret['obs_score'] + \
                            ret['rune_score']
    
    return ret
```

```python
match = recent_1000[0]
#for match in recent_1000:
players = match_df[match_df['match_id']==match].sort_values(['isRadiant','account_id']).reset_index(drop=True)

radiant = []
dire = []
for i, player in players.iterrows():
    last_25 = match_df[(match_df['match_id']<match) & (match_df['account_id']==player['account_id'])]
    fant = make_fantasy(last_25)
    if player['isRadiant']:
        radiant.append(fant['fantasy_score'].round(2))
    else:
        dire.append(fant['fantasy_score'].round(2))
print(dire)
```

```python
def get_match_data(matches):
    match_data = []
    errors = []
    for match in tqdm(matches,total=len(matches)):
        try:
            request = requests.request('GET',f"https://api.opendota.com/api/matches/{match}?api_key=c4b80265-46ad-4728-b0f4-ae0c1a175bef")
            data = request.json()
            if(data['players']):
                match_data.append(data)
        except:
            errors.append(match)
    
    df = pd.DataFrame(match_data)
    
    return df, errors
    
```
