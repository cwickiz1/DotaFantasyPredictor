```python
import pandas as pd
```

```python
def fantasy(df):
    """
    Kills = 0.3 * Kills
    Deaths = 5 - (0.3 * Deaths)
    Last Hits = 0.003 * Last Hit/Deny
    GPM = 0.002 * GPM
    Towers Killed = Towers Killed
    Roshans Killed = Roshans Killed
    Teamfights = 5 * Teamfight Participation Percentage
    Wards Placed = 0.5 * Obs Wards Placed
    Camps Stacked = 0.5 * Camps Stacked
    Runes = 0.25 * Runes Picked Up or Bottled
    First Blood = 4 if player got killing blow
    Stuns = 0.05 * Seconds of Stun
    """
    fantasy_df = pd.DataFrame()
    fantasy_df['account_id'] = df['account_id']
    fantasy_df['kill_points'] = df['kills'] * 0.3
    fantasy_df['death_points'] = 3 - (df['deaths'] * 0.3)
    fantasy_df['creep_score_points'] = (df['last_hits'] + df['denies']) * 0.003
    fantasy_df['gpm_points'] = df['gold_per_min'] * 0.002
    fantasy_df['tower_kill_points'] = df['towers_killed']
    fantasy_df['roshan_kill_points'] = df['roshans_killed']
    fantasy_df['teamfight_points'] = df['teamfight_participation'] * 3
    fantasy_df['obs_wards_planted_points'] = df['obs_placed'] * 0.5
    fantasy_df['camps_stacked_points'] = df['camps_stacked'] * 0.5
    fantasy_df['runes_grabbed_points'] = df['rune_pickups'] * 0.25
    fantasy_df['fb_points'] = df['firstblood_claimed'] * 4
    fantasy_df['stun_points'] = df['stuns'] * 0.05
    fantasy_df['total_fantasy_points'] = fantasy_df['kill_points'] + fantasy_df['death_points'] + fantasy_df['creep_score_points'] + fantasy_df['gpm_points'] + fantasy_df['tower_kill_points'] + fantasy_df['roshan_kill_points'] + fantasy_df['teamfight_points'] + fantasy_df['obs_wards_planted_points'] + fantasy_df['camps_stacked_points'] + fantasy_df['runes_grabbed_points'] + fantasy_df['fb_points'] + fantasy_df['stun_points']
    
    return fantasy_df
```

```python
df = pd.read_csv('playerMatchStats.csv')
```

```python
fantasy_df = fantasy(df)
```
